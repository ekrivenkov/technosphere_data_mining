Далее необходимо ввести новые признаки. Для каждого пользователя предлагается ввести следующие признаки:
- name_words - количество слов в имени
- screen_name_length - количество символов в псевдониме
- description_length - длина описания
- created_year - год создания аккаунта
- country_code - код страны
- verified - предлагается перевести в тип int

```python
import string

def create_new_features(df_users, features):
    # Introduce new features
    new_features = ["name_words", "screen_name_length", "description_length", "created_year", "country_code", "verified"]
    df_users["name_words"] = [x.count(" ") + 1 for x in df_users["name"]]
    df_users["screen_name_length"] = [len(x) for x in df_users["screen_name"]]
    df_users["description_length"] = [len(x) if not pd.isnull(x) else 0 for x in df_users["description"]]
    df_users["created_year"] = [x.year for x in df_users["created_at"]]
    df_users["verified"] = [int(x) for x in df_users["verified"]]
    
    country_code = {}
    i = 0
    for x in df_users["country"]:
        if x not in country_code:
            country_code[x] = i
            i += 1
          
    df_users["country_code"] = [country_code[x] for x in df_users["country"]]

    features = features + new_features
    
    return df_users, features
```

```python
features = ["lat", "lon", "followers_count", "friends_count", "statuses_count", "favourites_count", "listed_count"]
df_users, features = create_new_features(df_users, features)

x = df_users[pd.notnull(df_users.cls)][features].values
y = df_users[pd.notnull(df_users.cls)]["cls"].values
```

Необходимо реализовать функцию find_correlated_features, в которой нужно рассчитать коэффициенты корелляции и вывести те, которые больше 0.2. Подсказка: предлагается найти необходимую функцию в библиотеке np и реализовать find_correlated_features с использованием не более 5 строк кода (включая заголовок функции). (2 балла)

```python
def find_correlated_features(x, features):
    for i, feature_i in enumerate(features):
        for j, feature_j in enumerate(features):
            if i < j and abs(np.corrcoef(x[:, i], x[:, j])[0, 1]) > 0.2:
                print "Correlated features: %s + %s -> %.2f" % (feature_i, feature_j, np.corrcoef(x[:, i], x[:, j])[0, 1])
    
```

Построим взаимные распределения пар признаков в каждой из групп, а также гистограмму значений каждого из признаков с учетом целевой переменной.

```python
def plot_two_features_scatter(x_i, x_j, y):
    pl.scatter(x_j, x_i, c = ["red" if x else "green" for x in y], s = 4, alpha = 0.6,  edgecolors='none')
    
def plot_feature_histogram(x_i, y):
    pl.hist([[x for x, cl in np.nditer([x_i, y]) if cl], [x for x, cl in np.nditer([x_i, y]) if not cl]], 
            color = ["red", "green"], fill = True, alpha = 0.6)

def plot_dataset(x, y, features):
    num = len(features)
    fig = pl.figure(figsize = (15,15))
    for i, feature_i in enumerate(features):
        for j, feature_j in enumerate(features):
            fig.add_subplot(num, num, i * num + j + 1, xticklabels = [], yticklabels = [])
            
            if not i:
                pl.title(feature_j)
            if not j:
                pl.ylabel(feature_i)
                
            if i != j:
                plot_two_features_scatter(x[:, i], x[:, j], y)
            else:
                plot_feature_histogram(x[:, i], y)
                
    
    pl.show()
```
